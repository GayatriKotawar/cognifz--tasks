```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
import numpy as np
def categorize_ratings(y):
    return np.where(y < 3.0, "Low",
           np.where(y < 4.0, "Medium", "High"))
def round_predictions(y_pred):
    return np.round(y_pred * 10) / 10 
val_y_class = categorize_ratings(val_y) 
lin_pred_class = categorize_ratings(round_predictions(lin_pred)) 
tree_pred_class = categorize_ratings(round_predictions(tree_pred))
forest_pred_class = categorize_ratings(round_predictions(forest_pred))
def evaluate_classification(name, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted")
    recall = recall_score(y_true, y_pred, average="weighted")
    print(f" {name} Classification Performance:")
    print(f"   Accuracy: {acc:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall: {recall:.4f}\n")
evaluate_classification("Linear Regression", val_y_class, lin_pred_class)
evaluate_classification("Decision Tree", val_y_class, tree_pred_class)
evaluate_classification("Random Forest", val_y_class, forest_pred_class)

```
